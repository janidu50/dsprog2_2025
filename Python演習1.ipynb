### 問1

文字列"datascience"の文字を逆に（末尾から先頭に向かって）並べた文字列を出力してください

### 問2

「ヘエルスシトンニキア」という文字列の1,3,5,7,9文字目を取り出して連結した文字列を出力してください


```python
s2 = "ヘエルスシトンニキア"
# Python のインデックスは0始まりなので、1,3,5,7,9文字目を取り出すにはそれぞれ 0,2,4,6,8 を使う
chars = s2[0] + s2[2] + s2[4] + s2[6] + s2[8]
print(chars)  # 期待出力: ヘルシンキ
```

### 問3

「ヘルシンキ」＋「エストニア」の文字を先頭から交互に連結して文字列「ヘエルスシトンニキア」を出力してください


```python
s1 = "ヘルシンキ"
s2 = "エストニア"

result = ""

# 2つの文字列の長さに合わせて交互に取る
for a, b in zip(s1, s2):
    result += a + b

print(result)  # 期待出力: ヘエルスシトンニキア
```

### 問4

"Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成してください


```python
text = "Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"

# ① 単語に分解（空白でsplit）
words = text.split()

# ② アルファベット順にソート
sorted_words = sorted(words)

print(sorted_words)
```

### 問5

"It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."という文を単語に分解し，1, 2, 5, 6, 8, 9, 16番目の単語の先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）を格納した辞書を作成してください

```python
sentence = ("It is not the strongest of the species that survives, nor the most intelligent "
            "that survives. It is the one that is most adaptable to change.")

# ① 文を単語に分解
words = sentence.split()

# 必要な単語番号（1始まり）
indices = [1, 2, 5, 6, 8, 9, 16]

# ② 辞書にまとめる: {単語番号: 先頭2文字}
result_dict = {i: words[i - 1][:2] for i in indices}

print(result_dict)
```
